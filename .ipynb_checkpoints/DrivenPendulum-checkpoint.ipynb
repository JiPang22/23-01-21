```python
import matplotlib.pyplot as plt
from math import *

dt=1e-2;imax=10

fig, ax = plt.subplots(imax,2,figsize=(9,20))
plt.subplots_adjust(hspace=0.6)

for i in range(1,imax+1):
    tv=[];th1v=[];w1v=[]
    t=0;th1=0;w1=0;dw1=0;w=0.7;c=0.05
    f=i*0.1
    
    while(t<100):
        tv.append(t);th1v.append(th1);w1v.append(w1)

        dw1=-c*w1-sin(th1)+f*cos(w*t);dth1=w1

        t+=dt;th1+=dth1*dt;w1+=dw1*dt

        
    ax[i-1][0].plot(tv,w1v)
    ax[i-1][0].set_title(f'f={f}')
    ax[i-1][1].plot(th1v,w1v)
    ax[i-1][1].set_title(f'f={f}')
    
plt.savefig('plot.png')   

    
```
