```python
from math import *
import matplotlib.pyplot as plt
tv=[];th1v=[];th2v=[];w1v=[];w2v=[]
t=0;dt=1e-2;th1=0;th2=0;w1=0;w2=0;al1=0;al2=1
while t<100:
    tv.append(t);th1v.append(th1);th2v.append(th2);w1v.append(w1);w2v.append(w2)
    dw1=(1/2)*al2*cos(th1-th2)+(1/2)*w2**2*sin(th1-th2)-sin(th1)
    dw2=-al1*cos(th1-th2)+w1**2*sin(th1-th2)-sin(th2)
    dth1=w1
    dth2=w2
    t+=dt;th1+=dth1*dt;th2+=dth2*dt;w1+=dw1*dt;w2+=dw2*dt
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(tv,th2v)              # left-top
plt.xlabel('time (s)')
ax[0][1].plot(tv,w2v)               # right-top
ax[1][0].plot(th2v,w2v)            # left-bottom
ax[1][1].plot(th1v,w1v)         # right-bottom
```
